```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# T20 Matches Played
data = {
    'Player': [
        'AXAR PATEL','SURYA KUMAR YADAV','RISHAB PANT','ROHIT SHARMA',
        'AB DIVILLIERS','VIRAT KOHLI','SACHIN TENDULKAR','VIV RICHARDS',
        'MSD','HARDIK PANDYA','Mike Hesson','Amelia Kerr','Kane Williamson','Ross Taylor','Maxwell GJ','Steve smith',
         'Jaisurya','sangkara','rahul dravid','navjot singh siddhu' ],  
   
    'Matches': [10,19,30,45,60,80,100,120,70,55,29,65,67,56,44,39,34,54,76,44],
    'Runs': [250,300,1600,2200,2800,3400,4000,3600,1046,760,650,2098,1776,2096,3987,2043,2345,985,450,3400]
}

df = pd.DataFrame(data)

# Step 2: Features and Target
X = df[['Matches']]
y = df['Runs']

# Step 3: Train Model
model = LinearRegression()
model.fit(X, y)

# Step 4: Predictions
y_pred = model.predict(X)

# Step 5: Model Evaluation
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (RÂ²): {r2:.4f}")

# Step 6: Visualization
plt.scatter(X, y, color="blue", label="Actual Runs")
plt.plot(X, y_pred, color="red", label="Regression Line")
plt.xlabel("Matches Played")
plt.ylabel("Runs Scored")
plt.title("Player Performance Prediction (Matches vs Runs)")
plt.legend()
plt.show()

# Step 7: Predict Runs for a new player with 110 matches
new_data = pd.DataFrame({'Matches': [110]})  # â keep column name
predicted_runs = model.predict(new_data)

print(f"Predicted Runs for Yuvraj Singh with 110 matches: {predicted_runs[0]:.0f}")
```
